```python
# Collaborative Filtering where user must have reviewed 200 books and 50 is the average books
```

```python
user_with_rating = ratings_with_name.merge(users, on = 'User-ID')
```

```python
x = user_with_rating.groupby('User-ID').count()['Book-Rating']>= 200
user_with_most_review =x[x].index
user_with_most_review
```

```python
filter_1 = user_with_rating[user_with_rating['User-ID'].isin(user_with_most_review)]
filter_1
```

```python
y = filter_1.groupby('Book-Title').count()['Book-Rating']>=50
books_with_most_review = y[y].index
```

```python
filter_2 =filter_1[filter_1['Book-Title'].isin(books_with_most_review)]
filter_2
```

```python
pt = filter_2.pivot_table(index = 'Book-Title',
                     columns = 'User-ID',
                     values = 'Book-Rating')
```

```python
pt.fillna(0, inplace = True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
def recommendor(book_name):
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_score[index])), key= lambda x:x[1], reverse = True)[1:6]

  for i in similar_items:
    print(pt.index[i[0]])
```

```python
recommendor('Zoya')
```
